```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
cosine_similarity(data[0][0],data[0][0+1])
```

```python
for j in range(len(data)):
    similarity_values=[cosine_similarity(data[j][i],data[j][i+1]) for i in range(len(data[1])-1)]
    plt.plot(similarity_values)
    plt.show()
```
